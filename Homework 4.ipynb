#### b. Find the mean of all the numeric variables stratified by sex and age group where age groups are defined as ranges of 10 years (i.e. 0-10, 10-20, 20-30, etc).

```python
def create_age_group(x):
    floor_val = math.floor(x/10) * 10
    roof_val = math.ceil(x/10) * 10
    if (floor_val == roof_val):
        roof_val += 10
    return str(floor_val) + "-" + str(roof_val)
vcreate_age_group = np.vectorize(create_age_group)

HELPrct_sub = HELPrct
HELPrct_sub['age_group'] = vcreate_age_group(HELPrct_sub.age)
HELPrct_sub = HELPrct_sub.groupby(['sex', 'age_group']).mean(numeric_only=True).reset_index()
HELPrct_sub
```

### Question 2: The team IDs corresponding to Brooklyn baseball teams from the Teams data frame from the Lahman package are listed below. Use map int() to find the number of seasons in which each of those teams played by calling a function called count seasons.

```python
Teams.head().iloc[:, : 6]
```

```python
bk_teams = ["BR1", "BR2", "BR3", "BR4", "BRO", "BRP", "BRF"]
def count_seasons(ID):
    Teams_fn = Teams.loc[Teams.teamID == ID]
    Teams_fn = Teams_fn.groupby(Teams.teamID).agg('size').reset_index()
    Teams_fn = Teams_fn.rename(columns={0: 'count'})
    count = Teams_fn.iloc[0]['count']
    return count
total_seasons = list(map(count_seasons,bk_teams))
total_seasons
```

```python
def print_data(x, y):
    return x + " has: " + str(y) + " total season(s)"
list(map(print_data,bk_teams,total_seasons))
```
