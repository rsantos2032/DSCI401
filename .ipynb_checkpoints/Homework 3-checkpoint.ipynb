```python
# Loading the Dataset
import pandas as pd
Batting = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/Batting.csv")
Pitching = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/Pitching.csv")
People = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/People.csv")
```

### Question 1: Use the Batting, Pitching, and People tables in the Lahman package to answer the following questions:

#### a. Name every player in baseball history who has accumulated at least 300 home runs (HR) AND at least 300 stolen bases (SB). You can find the first and last name of the player in the People data frame. Join this to your result along with the total home runs and total bases stolen for each of these elite players.

```python
Batting_ = Batting.groupby('playerID').agg({'SB':'sum','HR':'sum'}).reset_index()
Batting_ = Batting_.loc[(Batting_['HR'] >= 300) & (Batting_['SB'] >= 300)]
Batting_ = pd.merge(Batting_, People, left_on = "playerID", right_on = "playerID", how = "left")
Batting_ = Batting_.loc[:, ["playerID", "nameFirst", "nameLast", "SB", "HR"]]
Batting_
```

#### b. Similarly, name every pitcher in baseball history who has accumulated at least 300 wins (W) and at least 3,000 strikeouts (SO).

```python
Pitching_ = Pitching.groupby('playerID').agg({'W':'sum','SO':'sum'}).reset_index()
Pitching_ = Pitching_.loc[(Pitching_['W'] >= 300) & (Pitching_['SO'] >= 3000)]
Pitching_ = pd.merge(Pitching_, People, left_on = 'playerID', right_on = 'playerID', how = 'left')
Pitching_ = Pitching_.loc[:, ['playerID', 'nameFirst', 'nameLast', 'W', 'SO']]
Pitching_
```

#### c. Identify the name and year of every player who has hit at least 50 home runs in a single season. Which player had the lowest batting average in that season?

```python
Batting_ = Batting.groupby(['playerID', 'yearID']).sum().reset_index()
Batting_['BA'] = Batting_['H']/Batting_['AB']
Batting_ = Batting_.loc[Batting_['HR'] >= 50]
Batting_ = pd.merge(Batting_, People, left_on = 'playerID', right_on = 'playerID', how = 'left')
Batting_ = Batting_.loc[:, ['playerID', 'yearID', 'HR', 'BA', 'nameFirst', 'nameLast']]
Batting_.sort_values(by = 'BA', ascending = True).head()
```

From the dataset, it appears that Pete Alonso had the lowest batting average in the 2019 season.

### Question 2: Use the nycflights13 package and the flights and planes tables to answer the following questions:

```python
planes = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/planes.csv")
flights = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/flights.csv")
```

#### a. What is the oldest plane (specified by the tailnum variable) that flew from New York City airports in 2013?

```python
planes_ = planes.rename(columns={'year': 'created'})
planes_['age'] = 2013 - planes_['created']
planes_ = pd.merge(planes_, flights, left_on = 'tailnum', right_on = 'tailnum', how = 'left')
planes_ = planes_.loc[:, ['tailnum', 'created', 'age']]
planes_ = planes_.sort_values(by = 'age', ascending = False)
planes_.head(1)
```

The oldest plane by tailnum is N381AA, which was created in 1956 and was 57 in 2013.

#### b. How many airplanes that flew from New York City are included in the planes table?

```python
flights_ = pd.merge(flights, planes, left_on = 'tailnum', right_on = 'tailnum', how = 'inner')
flights_ = flights_.groupby('tailnum').sum().reset_index()
len(flights_)
```

There are 3322 airplanes from the flights table that are included in the planes table.
